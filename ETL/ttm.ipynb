```python
import pandas as pd
import numpy as np
```

```python
df_report = pd.read_parquet(r'D:\python\financial statement prj\chatbot_financial_statement\csv\\financial_statement_v3.parquet')
df_report
```

```python
financial_report =df_report.dropna(subset=['category_code'])
```

```python
df_sort = financial_report.sort_values(by=['stock_code', 'category_code', 'year', 'quarter'],ascending=False).reset_index(drop=True)
df_sort = df_sort[df_sort['quarter'] != 0]
df_sort['sum_ttm']=df_sort.groupby(['stock_code','category_code'])['data'].apply(lambda x: x+x.shift(1,fill_value=0 )+x.shift(2,fill_value = 0)+ x.shift(3,fill_value = 0)).reset_index(drop = True)
df_sort

```
