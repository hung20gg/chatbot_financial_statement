```python
df_bank = pd.read_parquet('../csv/bank_financial_report_v2.parquet')
pivot_df  = df_bank[df_bank['category_code'].isin(['BS_212', 'BS_213'])].pivot_table(
    index=["stock_code", "year", "quarter"], 
    columns="category_code", 
    values="data"
).reset_index()

pivot_df['BS_211.1'] = pivot_df['BS_212'] + pivot_df['BS_213']

new_rows = pivot_df.melt(
    id_vars=["stock_code", "year", "quarter"], 
    value_vars=["BS_211.1"], 
    var_name="category_code", 
    value_name="data"
)

df_bank = pd.concat([df_bank, new_rows], ignore_index=True)
```

```python
pivot_df  = df_bank[df_bank['category_code'].isin(['BS_310', 'BS_320', 'BS_321', 'BS_322', 'BS_330', 'BS_340', 'BS_350', 'BS_360'])].pivot_table(
    index=["stock_code", "year", "quarter"], 
    columns="category_code", 
    values="data"
).reset_index()

pivot_df['BS_361'] = pivot_df['BS_310'] + pivot_df['BS_320'] + pivot_df['BS_321'] + pivot_df['BS_322'] + pivot_df['BS_330'] + pivot_df['BS_340'] + pivot_df['BS_350'] + pivot_df['BS_360']

new_rows = pivot_df.melt(
    id_vars=["stock_code", "year", "quarter"], 
    value_vars=["BS_361"], 
    var_name="category_code", 
    value_name="data"
)

df_bank = pd.concat([df_bank, new_rows], ignore_index=True)
df_bank.to_parquet('../csv/bank_financial_report_v2.parquet', index=False)
```

```python
df_sec = pd.read_parquet('../csv/securities_financial_report_v2.parquet')
pivot_df = df_sec[df_sec['category_code'].isin(['IS_020', 'IS_040'])].pivot_table(
    index=["stock_code", "year", "quarter"], 
    columns="category_code", 
    values="data"
).reset_index()

pivot_df['IS_040.1'] = pivot_df['IS_020'] - pivot_df['IS_040']

new_rows = pivot_df.melt(
    id_vars=["stock_code", "year", "quarter"], 
    value_vars=["IS_040.1"], 
    var_name="category_code", 
    value_name="data"
)

df_sec = pd.concat([df_sec, new_rows], ignore_index=True)
df_sec.to_parquet('../csv/securities_financial_report_v2.parquet', index=False)
```

```python
quarter_to_month = {
    0: 12,  # Quarter 0 is December of the same year
    1: 3,
    2: 6,
    3: 9,
    4: 12
}
```

```python
df_sec['date_added'] = pd.to_datetime(df_sec.apply(lambda row: f"{row['year']}-{quarter_to_month[row['quarter']]}-30", axis=1))
df_bank['date_added'] = pd.to_datetime(df_bank.apply(lambda row: f"{row['year']}-{quarter_to_month[row['quarter']]}-30", axis=1))
df_corp['date_added'] = pd.to_datetime(df_corp.apply(lambda row: f"{row['year']}-{quarter_to_month[row['quarter']]}-30", axis=1))
df_fs['date_added'] = pd.to_datetime(df_fs.apply(lambda row: f"{row['year']}-{quarter_to_month[row['quarter']]}-30", axis=1))
```

```python
df_sec.to_parquet('../csv/securities_financial_report_v2.parquet', index=False)
df_bank.to_parquet('../csv/bank_financial_report_v2.parquet', index=False)
df_corp.to_parquet('../csv/non_bank_financial_report_v2.parquet', index=False)
df_fs.to_parquet('../csv/financial_statement_v2.parquet', index=False)
```
