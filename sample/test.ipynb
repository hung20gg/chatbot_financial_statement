```python
from vnstock3 import Vnstock
import pandas as pd
```

```python
random_vn30 = ["TCB", "VCB", "VNM", "VIC", "VHM", "VRE", "BID", "CTG", "FPT", "HPG", "MBB", "MSN", "MWG" ,"SSI", "STB", "TPB", "VJC", "VPB", "VJC","VIB", "ACB","BCM","HDB","SHB"]
big_df = []
```

```python
for symbol in random_vn30:
    stock = Vnstock().stock(symbol=symbol, source='VCI')
    df = stock.finance.income_statement(period='quater', lang='en')
    df = df.loc[:,~df.columns.duplicated()].copy()
    
    columns = df.columns
    
    tickers = []
    yearReports = []
    lengthReports = []
    categories = []
    dataReports = []

    for index, data in df.iterrows():

        for column in columns:
            if column == 'ticker' or column == 'yearReport' or column == 'lengthReport':
                continue
            tickers.append(data['ticker'])
            yearReports.append(data['yearReport'])
            lengthReports.append(data['lengthReport'])
            categories.append(column)
            dataReports.append(data[column])
            
    df_transform = pd.DataFrame({'ticker': tickers, 'yearReport': yearReports, 'lengthReport': lengthReports, 'category': categories, 'data': dataReports})
    big_df.append(df_transform)
big_df = pd.concat(big_df)
big_df.to_csv('data.csv', index=False)
```
