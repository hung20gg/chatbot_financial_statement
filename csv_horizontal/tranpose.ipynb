```python
import pandas as pd
import numpy as np
```

## Securities report 

```python
sec_report = pd.read_parquet(r'D:\python\financial statement prj\chatbot_financial_statement\csv\v3\securities_financial_report.parquet')
sec_report
```

```python
sec_report_hori = sec_report.pivot_table(
    index=['stock_code', 'year', 'quarter'],  # Set these as unique identifiers
    columns='category_code',                 # Each category_code becomes a column
    values='data',                           # Values to fill in the table
    aggfunc='first'                          # Take the first occurrence if duplicates exist
).reset_index()

# Flatten column names
sec_report_hori.columns.name = None  # Remove column index name
sec_report_hori.columns = [str(col) for col in sec_report_hori.columns]
```

```python
sec_report_hori.to_csv('..\csv_horizontal\sec_financial_report_hori.csv', index=False)
```

# industry

```python
industry = pd.read_parquet(r'D:\python\financial statement prj\chatbot_financial_statement\csv\industry_report_v3.parquet')
industry.head()
```

```python
industry_report_hori = industry.pivot_table(
    index=['industry', 'year', 'quarter'],  # Set these as unique identifiers
    columns='category_code',                 # Each category_code becomes a column
    values=['data_sum', 'data_mean'],        # Values to fill in the table
    aggfunc='first'                          # Take the first occurrence if duplicates exist
).reset_index()

# Flatten the hierarchical column names except for 'industry', 'year', and 'quarter'
industry_report_hori.columns.name = None  # Remove column index name
industry_report_hori.columns = [
    f"{col[0]}_{col[1]}" if col[1] else col[0] 
    for col in industry_report_hori.columns]

```

```python
industry_report_hori.head()
```

```python
industry_report_hori.to_csv(r'..\csv_horizontal\industry_financial_report_hori.csv', index=False)
```
