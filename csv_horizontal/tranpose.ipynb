```python
import pandas as pd
import numpy as np
```

## Securities report 

```python
sec_report = pd.read_csv('..\csv\securities_financial_report_v2_2.csv')
sec_report
```

```python
sec_report_hori = sec_report.pivot_table(
    index=['stock_code', 'year', 'quarter'],  # Set these as unique identifiers
    columns='category_code',                 # Each category_code becomes a column
    values='data',                           # Values to fill in the table
    aggfunc='first'                          # Take the first occurrence if duplicates exist
).reset_index()

# Flatten column names
sec_report_hori.columns.name = None  # Remove column index name
sec_report_hori.columns = [str(col) for col in sec_report_hori.columns]
```
